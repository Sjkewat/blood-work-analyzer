```python
with open("blood_work.txt" , "r")as f:
    blood_report = f.read()

print(blood_report[:200])
```

```python
extraction_prompt = f"""
You are a medical data extraction assistant .

from the blood report below , extract All test values and classify each as HIGH , LOW , or NORMAL 
based on the refrance ranges provide in the report .

format your response as:
-  Test Name: Value  | status : HIGH/LOW/NORMAL | refrance : range

BLood report:
{blood_report}
"""
from dotenv import load_dotenv
from langchain_google_genai import ChatGoogleGenerativeAI

load_dotenv()

llm = ChatGoogleGenerativeAI(model ="gemini-2.5-flash")
extraction_response = llm.invoke(extraction_prompt)
extracted_values= extraction_response.text

print("=== STAGE 1 : EXTRACTED VALUES ===")
print(extracted_values)
```
