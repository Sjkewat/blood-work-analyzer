```python
from dotenv import load_dotenv
from langchain_google_genai import ChatGoogleGenerativeAI

load_dotenv()
```

```python
llm = ChatGoogleGenerativeAI(model ="gemma-4-31b-it")

response = llm.invoke("How many moon Jupitar have")
print(response.text)
```
